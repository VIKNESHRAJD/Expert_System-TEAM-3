```python
import tkinter as tk

class Calculator:
    def __init__(self, master):
        self.master = master
        master.title("Simple Calculator")

        # create display widget
        self.display = tk.Entry(master, width=25, justify='right', font=('Arial', 16))
        self.display.grid(row=0, column=0, columnspan=4, padx=5, pady=5)

        # create number buttons
        button_texts = ['7', '8', '9', '4', '5', '6', '1', '2', '3', '0', '.', 'C']
        self.buttons = []
        for i, text in enumerate(button_texts):
            button = tk.Button(master, text=text, width=5, font=('Arial', 14), command=lambda text=text: self.button_click(text))
            row, col = divmod(i, 3)
            button.grid(row=row+1, column=col, padx=5, pady=5)
            self.buttons.append(button)

        # create operator buttons
        self.operator_buttons = []
        for i, text in enumerate(['/', '*', '-', '+']):
            button = tk.Button(master, text=text, width=5, font=('Arial', 14), command=lambda text=text: self.operator_click(text))
            button.grid(row=i+1, column=3, padx=5, pady=5)
            self.operator_buttons.append(button)

        # create equals button
        self.equals_button = tk.Button(master, text='=', width=5, font=('Arial', 14), command=self.equals_click)
        self.equals_button.grid(row=4, column=2, padx=5, pady=5)

        # create clear button
        self.clear_button = tk.Button(master, text='Clear', width=5, font=('Arial', 14), command=self.clear_click)
        self.clear_button.grid(row=4, column=0, padx=5, pady=5, columnspan=2)

        # initialize state variables
        self.operator = None
        self.num1 = None
        self.num2 = None

    def button_click(self, text):
        current = self.display.get()
        if text == '.':
            if '.' in current:
                return
            elif current == '':
                self.display.insert(0, '0.')
                return
        self.display.delete(0, tk.END)
        self.display.insert(0, current+text)

    def operator_click(self, text):
        self.operator = text
        self.num1 = float(self.display.get())
        self.display.delete(0, tk.END)

    def equals_click(self):
        if self.operator is None:
            return
        self.num2 = float(self.display.get())
        if self.operator == '/':
            result = self.num1 / self.num2
        elif self.operator == '*':
            result = self.num1 * self.num2
        elif self.operator == '-':
            result = self.num1 - self.num2
        elif self.operator == '+':
            result = self.num1 + self.num2
        self.display.delete(0, tk.END)
        self.display.insert(0, result)
        self.operator = None

    def clear_click(self):
        self.display.delete(0, tk.END)
        self.operator = None
        self.num1 = None
        self.num2 = None


root = tk.Tk()
calculator = Calculator(root)
root.mainloop()
```
